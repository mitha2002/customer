```python
import matplotlib.pyplot as plt
```

```python
products=["mobile","system","car","bike","auto"]
price=[10000,40000,60000,30000,50000]
```

```python
plt.plot(products,price)
plt.plot(products,price)
plt.show()
```

```python
plt.bar(products,price)
plt.show()
```

```python
colors=['red','yellow','brown','green','blue']
plt.bar(products,price,colors=colors,alpha=0.4,edgecolor='black',linewidth=4,linestyle=':',label='tvs')
plt.grid(axis='x',linestyle="-.")
plt.xlabel("products")
plt.ylabel("price")
plt.title("shop")
plt.legend()
plt.show()
```
