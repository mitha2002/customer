```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
x=[1,2,3,3,5,6]
y=[1,2,2,4,5,6]
z=[1,2,5,4,6,6]
ax.plot3D(x,y,z,color='green')
ax.set_title("3d")
plt.show()
```

```python
food=["itly","pongal","vada","dhosa","poori","veg briyani"]
price=[20,40,25,50,60,80]
```

```python
fig,ax=plt.subplots(2,2,figsize=(10,8))

ax[0][0].plot(food,price,color='green')
ax[0][0].set_title("Line plot")

ax[1][0].bar(food,price,color='grey')
ax[1][0].set_title("Bar plot")

ax[0][1].scatter(food,price,color='brown')
ax[0][1].set_title("Scatter plot")

ax[1][1].pie(price,labels=food,autopct="%1.0f%%",startangle=90)
ax[1][1].set_title("pie plot")
ax[1][1].axis("equal")
plt.tight_layout()
plt.show()
```
