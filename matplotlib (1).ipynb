```python
import matplotlib.pyplot as plt
```

```python
products=["mobile","system","car","bike","auto"]
price=[10000,40000,60000,30000,50000]
```

```python
plt.plot(products,price)
plt.plot(products,price)
plt.show()
```

```python
plt.bar(products,price)
plt.show()
```

```python
colors=['red','yellow','brown','green','blue']
plt.bar(products,price,color=colors,alpha=0.4,edgecolor='black',linewidth=4,linestyle=':',label="tvs")
plt.grid(axis='x',linestyle="-.")
plt.xlabel("products")
plt.ylabel("price")
plt.title("shop")
plt.legend()
plt.show()
```

```python
plt.pie(price,labels=products,autopct='%1.0f%%')
plt.show()
```

```python
plt.plot(products,price,color='black',linewidth=0.8,linestyle='-.',marker='*',label="pricetag")
plt.legend()
plt.show()
```
